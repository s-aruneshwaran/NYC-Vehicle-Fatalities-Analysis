**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/drive/MyDrive/TDSP Explorer dataset/Motor_Vehicle_Collisions_-_Crashes_20250211.csv"
data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=average_crashes_per_hour.index, y=average_crashes_per_hour.values)
plt.title('Average Number of crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')
plt.xticks(range(0, 24))
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes.index, daily_crashes.values, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date')
plt.ylabel('Number of Crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

**Step 3:** Next, we will begin "Severity Mapping." We'll plot crashes on the map and code them based on severity, distinguishing between crashes that resulted in injuries and those that led to fatalities. This will give us a visual representation of where severe crashes tend to occur. </br>

You may initially want to code these incidents by using different colors, but it's important to think about your map and how accessible it is. Will a color-coded map be easy to read for everyone? Let's make a map that's inclusive for people with color blindness by also creating differently-shaped markers (squares, circles, and triangles) for crashes, injuries, and fatalities.

````python
#TODO: Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "Red"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "Yellow"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "Green"  # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

````python
# Filter for crashes with at least one fatality
fatal_crashes = data[data['NUMBER OF PERSONS KILLED'] > 0]

# Drop rows where 'BOROUGH' is missing
fatal_crashes = fatal_crashes.dropna(subset=['BOROUGH'])

print("Number of fatal crashes in the dataset:", len(fatal_crashes))

````

````python
# Visualization 1: Fatalities by Borough

# Group by borough and sum the number of persons killed
fatalities_by_borough = fatal_crashes.groupby('BOROUGH')['NUMBER OF PERSONS KILLED'].sum().sort_values(ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=fatalities_by_borough.index, y=fatalities_by_borough.values, palette="viridis")
plt.title('Total Fatalities by Borough (10/2014 - 12/2022)')
plt.xlabel('Borough')
plt.ylabel('Number of Fatalities')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print the top fatalities
print(fatalities_by_borough)
````

This analysis clearly shows a significant disparity in traffic fatalities across New York City's boroughs between late 2014 and the end of 2022.

>* Key Insights from the Data
Brooklyn and Queens are the Most Dangerous: Brooklyn is unequivocally the borough with the most traffic fatalities, recording 669 deaths in this period. Queens is second with 558 deaths. Together, these two boroughs account for over 60% of all traffic fatalities in NYC, making them the most critical areas for safety interventions.

>* Staten Island is the Least Dangerous: On the other end of the spectrum, Staten Island has significantly fewer fatalities (103 deaths) than any other borough. In fact, Brooklyn has over six times more fatalities than Staten Island.

>* A Clear Hierarchy Exists: The ranking is distinct, with a clear drop-off in fatalities from Brooklyn and Queens to Manhattan (363), the Bronx (300), and finally Staten Island.

````python
# Visualization 2: Top Contributing Factors in Fatal Crashes by Borough

# Function to get the top contributing factors for a given borough
def get_top_factors(borough_name):
    borough_fatal_crashes = fatal_crashes[fatal_crashes['BOROUGH'] == borough_name]
    top_factors = borough_fatal_crashes['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(5)
    return top_factors

# Get the top factors for each borough
top_factors_manhattan = get_top_factors('MANHATTAN')
top_factors_brooklyn = get_top_factors('BROOKLYN')
top_factors_queens = get_top_factors('QUEENS')
top_factors_bronx = get_top_factors('BRONX')
top_factors_staten_island = get_top_factors('STATEN ISLAND')

# Combine the data for plotting
all_top_factors = pd.concat([top_factors_manhattan, top_factors_brooklyn, top_factors_queens, top_factors_bronx, top_factors_staten_island], axis=1)
all_top_factors.columns = ['MANHATTAN', 'BROOKLYN', 'QUEENS', 'BRONX', 'STATEN ISLAND']
all_top_factors = all_top_factors.fillna(0) # Fill NaN with 0 for factors not in a specific borough

# Create a grouped bar chart
all_top_factors.T.plot(kind='bar', figsize=(15, 8), width=0.8)
plt.title('Top 5 Contributing Factors in Fatal Crashes by Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Fatal Crashes')
plt.xticks(rotation=45)
plt.legend(title='Contributing Factor', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

Findings from this analysis,

>* Decoding the Dangers: A Borough-by-Borough Analysis of Fatal Crashes
This analysis reveals that while tragic, fatal crashes are not random. Each NYC borough has a distinct "crash signature," with different behaviors causing the most harm. Understanding these unique patterns is the key to creating effective, targeted safety solutions.

>* The Overarching Challenge: A Lack of Specifics
Before diving into the known causes, the most significant finding is the prevalence of the "Unspecified" category. In the four most populous boroughs, it's the single largest bar, meaning the primary reason for a vast number of fatal crashes is officially unknown. This highlights a critical need for more rigorous and consistent data collection at crash scenes to better understand and prevent future tragedies.

>* The Unique "Crash Signature" of Each Borough
When we examine the known causes, clear patterns emerge:

>* Brooklyn & The Bronx - The Inattention Problem: The leading known causes of fatal crashes in Brooklyn and the Bronx are consistently Driver Inattention/Distraction and Failure to Yield Right-of-Way. This points to a dangerous combination of distracted drivers and conflicts in dense traffic and at intersections.

>* Queens - The Intersection Violation Problem: Queens stands out with a clear focus on dangerous actions at intersections. The top two known causes are Failure to Yield Right-of-Way and Traffic Control Disregarded (e.g., running red lights or stop signs). This suggests that blatant violations of traffic rules at intersections are a primary driver of fatalities there.

>* Manhattan - The Pedestrian Conflict Zone:  While driver errors are still high, Manhattan's unique signature is the significant role of Pedestrian/Bicyclist Error/Confusion. This reflects the borough's immense foot and bike traffic and underscores the constant, high-stakes conflict between vehicles and vulnerable road users in the city's densest environment.

>* Staten Island - The Speed Problem: Staten Island is a complete outlier. The number one known cause of fatal crashes is Unsafe Speed. This is fundamentally different from the other boroughs and indicates that fatal crashes there are more likely to occur not in dense gridlock, but on highways and wider roads where high, dangerous speeds are possible.

>* For Brooklyn & the Bronx: Focus on campaigns and enforcement targeting distracted driving.

>* For Queens: Prioritize enforcement of traffic controls at major intersections and consider engineering solutions to make them safer.

>* For Manhattan: Invest heavily in infrastructure that protects pedestrians and cyclists, such as expanded protected bike lanes and more prominent, safer crosswalks.

>* For Staten Island: Implement a dual-focused safety strategy.

>* Address Inattention: Launch awareness campaigns and initiatives targeting distracted driving, just as you would in the other boroughs.

>* Tackle Speed: Crucially, continue with targeted speed reduction strategies, including enforcement and traffic-calming road designs. The fact that speed is a top-two cause of death makes it an essential and distinguishing factor to address for that borough's safety.
